```python
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import json
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
data = pd.read_json('key.json', lines=True)
```

```python
speed = data['speed'][0]
angle = data['angle'][0]
```

```python
t = np.linspace(0, len(speed)-1,len(speed))

fig = plt.figure(figsize=(15,6))

ax1 = fig.add_subplot(111)
ax1.plot(t, angle,label = 'heading',marker='x',color='b')
ax1.set_ylabel('angle')
plt.legend(loc='upper left')


ax2 = ax1.twinx()  
ax2.plot(t, speed,label = 'speed',marker='o',color='r')
ax2.set_ylabel('speed')
ax2.set_xlabel('time')

plt.legend(loc='upper right')
plt.show()    
```
